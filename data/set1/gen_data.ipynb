```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = datasets.make_blobs(600, centers=2, cluster_std=1.5)
X1 = X[np.where(Y == 0)]
X2 = X[np.where(Y == 1)]

train_idx = np.random.choice(600, 500, replace=False)

X_train = X[train_idx]
Y_train = Y[train_idx]

mask = np.ones(600, dtype=bool)
mask[train_idx] = False

X_test = X[mask]
Y_test = Y[mask]
```

```python
plt.scatter(X1[:, 0], X1[:, 1])
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
np.savez_compressed('./all', X=X, Y=Y)
np.savez_compressed('./train', X=X_train, Y=Y_train)
np.savez_compressed('./test', X=X_test, Y=Y_test)
```
